```python
from pandas.conftest import axis_1
import pandas as pd

df = pd.read_csv('ALLFLOWMETER_HIKARI2021.csv')
df.head(10)
```

```python
object_columns = df.select_dtypes(include=['object']).columns
print(object_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in object_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
print(df['traffic_category'].unique())
```

```python
#Kolom yang dicek
column_to_check = "traffic_category"

# Hitung IQR
Q1 = df[column_to_check].quantile(0.25)
Q3 = df[column_to_check].quantile(0.75)
IQR = Q3 - Q1

# Hitung batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cek outlier
outliers = df[(df[column_to_check] < lower_bound) | (df[column_to_check] > upper_bound)]

if outliers.empty:
    print(f"Tidak ada outlier dalam kolom {column_to_check}.")
else:
    print(f"Ada outlier dalam kolom {column_to_check}.")
    print(outliers)
```

```python
from imblearn.over_sampling import SMOTE

# Memisahkan fitur dan target
X = df.drop(columns=['traffic_category'])
y = df['traffic_category']

# Menggunakan SMOTE untuk menyeimbangkan data
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Mengecek distribusi target setelah SMOTE
print(y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Menormalisasi fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np

# Mengubah data menjadi bentuk (samples, timesteps, features)
X_train_rnn = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_rnn = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout

# Membangun model RNN
model = Sequential()

# Menambahkan layer RNN
model.add(SimpleRNN(64, activation='relu', input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])))
model.add(Dropout(0.2))  # Dropout untuk mengurangi overfitting

# Menambahkan output layer
model.add(Dense(6, activation='softmax'))  # Gunakan 'softmax' untuk klasifikasi multi kelas

# Menyusun model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Menampilkan ringkasan model
model.summary()
```

```python
from sklearn.preprocessing import LabelEncoder

# Memetakan ulang kelas
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

# Memastikan kelas terkode dengan benar
print(np.unique(y_train_encoded))
```

```python
# Melatih model
history = model.fit(X_train_rnn, y_train_encoded, epochs=5, batch_size=32, validation_data=(X_test_rnn, y_test_encoded))
```

```python
# Evaluasi model pada data uji
loss, accuracy = model.evaluate(X_test_rnn, y_test_encoded)

# Menampilkan hasil evaluasi
print(f"Loss pada data uji: {loss}")
print(f"Akurasi pada data uji: {accuracy}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Melakukan prediksi pada data uji
y_pred = model.predict(X_test_rnn)
y_pred_classes = np.argmax(y_pred, axis=1)  # Mengambil kelas dengan probabilitas tertinggi

# Menampilkan classification report
print(classification_report(y_test_encoded, y_pred_classes))

# Membuat confusion matrix
cm = confusion_matrix(y_test_encoded, y_pred_classes)

# Menampilkan confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Mengambil nilai akurasi dan loss dari history
history_dict = history.history
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

# Visualisasi Akurasi
plt.figure(figsize=(12, 6))

# Plot Akurasi
plt.subplot(1, 2, 1)
plt.plot(acc, label='Akurasi Pelatihan')
plt.plot(val_acc, label='Akurasi Validasi')
plt.title('Akurasi Pelatihan dan Validasi')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(loss, label='Loss Pelatihan')
plt.plot(val_loss, label='Loss Validasi')
plt.title('Loss Pelatihan dan Validasi')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Tampilkan plot
plt.tight_layout()
plt.show()
```
